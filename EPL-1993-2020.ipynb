```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('EPL_Set.csv')
```

```python
df.head()
```

```python
df.drop("Div",axis=1,inplace=True)
```

```python
df.head()
```

```python
df.isnull()
```

```python
df.tail()
```

```python
df.dtypes
```

```python
# Changing datatypes

df["Season"] = df["Season"].astype('str')
df["HomeTeam"] = df["HomeTeam"].astype('category')
df["AwayTeam"] = df["AwayTeam"].astype('category')
df["FTHG"] = df["FTHG"].astype(int)
df["FTAG"] = df["FTAG"].astype(int)
df["FTR"] = df["FTR"].astype('category')
df["HTHG"] = df["HTHG"].astype(float)
df["HTAG"] = df["HTAG"].astype(float)
df["HTR"] = df["HTR"].astype('category')
```

```python
df.dtypes
```

```python
df["HomeTeam"].value_counts(dropna=False)
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#Plot bar chart to show amount of Home Games played since 1993 
plt.figure(figsize=(17, 10))
sns.countplot(y='HomeTeam',data=df)
plt.xticks(rotation='horizontal')
plt.xlabel('Number of Games Played at Home')
plt.title('Number of game played at home in the EPL from 1993 to 2018')
plt.show()
```
